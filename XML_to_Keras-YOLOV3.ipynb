```python
import os
import re
import xml.etree.ElementTree as ET
```

```python
os.getcwd()
a=os.listdir("C:/Users/colin.huang/Desktop/pets/downloads/cat")
b=os.listdir("C:/Users/colin.huang/Desktop/pets/downloads/dog")
x=[]
for i in a:
    if len(re.findall(".+xml",i)):
        x.append("C:/Users/colin.huang/Desktop/pets/downloads/cat"+"/"+i)
for i in b:
    if len(re.findall(".+xml",i)):
        x.append("C:/Users/colin.huang/Desktop/pets/downloads/dog"+"/"+i)

x
```

```python
final_end=[]
for j in x:
    tree=ET.parse(j)
    root=tree.getroot()
    end=[]
    for i in root:
        if i.tag=='path':   
            end.append(i.text)
        elif i.tag=='object':
            for con in i:
                if con.tag=='name':                
                    #end.append(con.text)
                    name=con.text
                elif con.tag=='bndbox':
                    count=1; 
                    for coordinare in con:
                        if(count%4==0):
                            end.append(coordinare.text)
                            end.append(name)
                        else:
                            end.append(coordinare.text)
                        count+=1
                     
    final_end.append(end)

final_end=[[w.replace("\\","/") for w in q] for q in final_end]
final_end
```

```python
f=open('C:/Users/colin.huang/Desktop/demo.txt','w')
for i in final_end:
    k=','.join([str(j) for j in i]) 
    k=k.replace(".jpg,",".jpg ")
    k=k.replace(".JPG,",".jpg ")
    k=k.replace(".png,",".png ")
    k=k.replace(".PNG,",".png ")
    k=k.replace(".jpeg,",".jpeg ")
    k=k.replace(",cat,",",0 ")
    k=k.replace(",cat",",0")
    k=k.replace(",dog,",",1 ")
    k=k.replace(",dog",",1")
    print(k)
    f.write(k+"\n")
f.close()
```
